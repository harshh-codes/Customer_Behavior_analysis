```python
import pandas as pd
import numpy as np
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.info()
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
## Feature Engineering
labels = ['Young','Adult','Middle_Aged','Senior_Citizen']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Annually': 365,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df = df.drop('promo_code_used' , axis=1)
```

```python
import sys
import psycopg2
from sqlalchemy import create_engine
import urllib.parse  # Standard library to handle special characters

# 1. Setup credentials
username = "postgres"
password = "Your_password"
host = "localhost"
port = "5432"
database = "customer_behavior"

# 2. URL-encode the password (this turns @ into %40)
safe_password = urllib.parse.quote_plus(password)

# 3. Create the engine with the safe password
engine = create_engine(f"postgresql+psycopg2://{username}:{safe_password}@{host}:{port}/{database}")

# 4. Load df into PostgreSQL
try:
    table_name = "customer"
    df.to_sql(table_name, engine, if_exists="replace", index=False)
    print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
except Exception as e:
    print(f"An error occurred: {e}")
```
